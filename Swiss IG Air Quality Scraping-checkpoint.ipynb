```python
import pandas as pd
import numpy as np
import sys

import time
from time import sleep

!conda install --yes --prefix {sys.prefix} selenium
from selenium import webdriver
from selenium.common.exceptions import NoSuchElementException
from selenium.webdriver.common.keys import Keys
```

```python
path = r'C:\Users\IMalik\Documents\ChromeDriver\chromedriver_win32\chromedriver.exe'

URL_Exten = ['basel-city/basel', 'bern', 'geneva/geneve', 'vaud/lausanne', 'ticino/lugano', 'lucerne', 'st-gallen/sankt-gallen', 'zurich']

def get_AI(Country, CityList, Verbose, slp_time):
    
    options = webdriver.ChromeOptions()
    driver = webdriver.Chrome(executable_path=path, options=options)
    
    AQI_List = []
    
    for City in CityList:
        Baseurl = "https://www.iqair.com/us/{country}/{city}"
        url = Baseurl.format(country=Country,city=City)
        driver.get(url)
        driver.maximize_window()
    
        time.sleep(slp_time)
    
        try:
            score = driver.find_element_by_xpath('.//*[@id="content-wrapper"]/app-routes-resolver/div/app-city/div[2]/div[2]/app-aqi-overview/div/div[1]/div').text
            
            if len(score.split())-1 == 7:
                
                AQ_Index = {
                    'City' : City,
                    'Air Quality Index' : score.split()[2],
                    'Air Quality Catagory' : score.split()[7]
                }    
            elif len(score.split())-1 == 6:
                AQ_Index = {
                    'City' : City,
                    'Air Quality Index' : score.split()[2],
                    'Air Quality Catagory' : score.split()[6]
                }
            
            #print(len(score.split())-1)
            AQI_List.append(AQ_Index)
            
        except NoSuchElementException:
            pass
    
    df = pd.DataFrame.from_dict(AQI_List)
    driver.close()
    return df
```
